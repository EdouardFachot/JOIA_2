```
import pandas as pd
import requests
from bs4 import BeautifulSoup
import csv

page_number=1
try:
    with open('categories.csv','w',newline="") as csv_file:

        csv_write = csv.writer(csv_file)
        csv_write.writerow(['Categories', 'Link'])
        for i in range(1):
            url = f'https://www.amazon.com/Best-Sellers/zgbs/ref=zg_bs_unv_t_0_t_1'

            header ={
                'user-agent': 'Mozilla/5.0 (compatible; Googlebot/2.1; +http://www.google.com/bot.html)'
            }

            content = requests.get(url,header).text
            soup = BeautifulSoup(content,'lxml')
            
            try:
              wrapper = soup.find('ul',id="zg_browseRoot")
            except Exception as e:
                    print(e)
            try:
              single = wrapper.find_all('a')
            except Exception as e:
                    print(e)
            try:
              for single in single:
                
                csv_write.writerow([single.text, single['href']])
            except Exception as e:
                    print(e)
            page_number += 1
except Exception as e:
    print(e)





categories = pd.read_csv('categories.csv')

try:
    with open('sub_categories.csv','w',newline="") as csv_file:

        csv_write = csv.writer(csv_file)
        csv_write.writerow(['Sub_Categories', 'Link', 'Category', 'Link_Category' ])
        for category in categories['Link'] : 
            url = category

            header ={
                'user-agent': 'Mozilla/5.0 (compatible; Googlebot/2.1; +http://www.google.com/bot.html)'
            }

            content = requests.get(url,header).text
            soup = BeautifulSoup(content,'lxml')
            
            try:
              wrapper = soup.find('ul',id="zg_browseRoot")
            except Exception as e:
                    print(e)
            try:
              upper_category = soup.find('span',class_="category").text
            except Exception as e:
                    print(e)
            try:
              single = wrapper.find_all('a')
            except Exception as e:
                    print(e)
            
            try:
              for single in single:
                
                csv_write.writerow([single.text, single['href'], upper_category, category])
            except Exception as e:
                    print(e)
            page_number += 1
except Exception as e:
    print(e)

```

```
import pandas as pd
import requests
from bs4 import BeautifulSoup
import csv

page_number=1
try:
    with open('best_sell.csv','w',newline="") as csv_file:

        csv_write = csv.writer(csv_file)
        csv_write.writerow(['Rank','Title','Price','Rating', 'Number of Reviews', 'Category'])
        for i in categories['Link']:
            print(i)
            url = i

            header ={
                'user-agent': 'Mozilla/5.0 (compatible; Googlebot/2.1; +http://www.google.com/bot.html)'
            }

            content = requests.get(url,header).text
            soup = BeautifulSoup(content,'lxml')
            

            wrapper = soup.find('ol',id="zg-ordered-list")
            upper_category = soup.find('span',class_="category").text

            single = wrapper.find_all('li',class_="zg-item-immersion")

            for single in single:
                    
                # Finding title of the product
                try:
                  title = single.find('span',class_="aok-inline-block zg-item").a.find_all('div')[1]
                  title =title.text
                  title = title.replace('\n',' ')
                  title = title.strip()
                except Exception as e:
                    print(e)
                # Finding the rank
                try:
                  rank = single.find('span',class_="zg-badge-text")
                  rank = rank.text[1:]
                except Exception as e:
                    print(e)

                try:
                    #Finding the price of the product
                    price = single.find('span',class_="aok-inline-block zg-item").find('span',class_="p13n-sc-price")
                    price = price.text
                    #price = price[1:]
                    price = price.replace(',','')
                except Exception as e:
                    print(e)
                

                try:
                    #Finding the price of the product
                    reviews = single.find('span',class_="aok-inline-block zg-item").find('a',class_="a-size-small a-link-normal")
                    reviews = reviews.text
                    #price = price[1:]
                    reviews = reviews.replace(',','')
                except Exception as e:
                    print(e)
                #Finding the rating of the product
                try:
                  rate = single.find('span',class_="aok-inline-block zg-item").i.text[0:3]
                except Exception as e:
                    print(e)
                csv_write.writerow([rank,title,price,rate, reviews, upper_category])
            page_number += 1
except Exception as e:
    print(e)
```
